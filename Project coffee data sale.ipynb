```python
#Q6 convert data to datetime
df['Date'] = pd.to_datetime(df['Date'])
#calculate Revenue(money is already revenue)
revenue_list = []
for i in range(len(df)):
    revenue_list.append(df.loc[i, 'money'])
df['Revenue'] = revenue_list

#calculate Revenue Growth
growth_list = [0]
for i in range(1, len(df)):
    growth = (df.loc[i, 'Revenue'] - df.loc[i-1, 'Revenue']) / df.loc[i-1, 'Revenue'] * 100
    growth_list.append(growth)
df['Revenue_Growth'] = growth_list

#calculate Total sale (same a money)
total_list = []
for i in range(len(df)):
    total_list.append(df.loc[i, 'money'])
df['Total_Sales'] = total_list

#calculate Percentage relative to highest sales
max_total = max(total_list)
percentage_list = []
for total in total_list:
    perc = total / max_total * 100
    percentage_list.append(perc)
df['Percentage'] = percentage_list

df

```

```python
#convert date to datetime
df['Date'] = pd.to_datetime(df['Date'])

#plot line foe each coffee name
for coffee in df['coffee_name'].unique():
    coffee_data = df[df['coffee_name'] == coffee]
    plt.plot(coffee_data['Date'], coffee_data['Revenue'], marker='o', label=coffee)

plt.title('Daily Revenue Comparison by Coffee name')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()

```
